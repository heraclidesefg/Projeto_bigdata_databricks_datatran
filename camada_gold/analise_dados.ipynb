Camada Gold - Análise de acidentes Rodoviários
Pyspark / Databricks

```python
from pyspark.sql.functions import *

#======================================
# Objetivo das Aálises
#======================================

# 1- Estados com maior número de acidentes
# 2- Meses com maior número de ocorrências de acidentes
# 3- Tipos de acidentes mais frequentes
```

```python
#===========================================
# Carregar tabela silver
#===========================================

df_gold = spark.table("projeto_profissional_bigdata.datatran_2018.silver_datatran18")
```

```python
#==========================================
# Visualizar dados
#==========================================
df_gold.show(5)
display(df_gold)
```

```python
#======================================
# Pergunta 1
# Estados com maior número de acidentes
#=======================================

gold_acidente_por_estado = (
    df_gold.groupBy("uf")
    .agg(count("*").alias("total_acidentes"))
    .orderBy(col("total_acidentes").desc())
)

print("ESTADOS COM MAIS ACIDENTES")
gold_acidente_por_estado.show(5)
```

```python
#====================================
# Salvar tabela gold estados
#====================================

gold_acidente_por_estado.write\
    .format("delta")\
    .mode("overwrite")\
    .saveAsTable("projeto_profissional_bigdata.datatran_2018.gold_acidente_por_estado")        
```

```python
%sql
SELECT * FROM projeto_profissional_bigdata.datatran_2018.gold_acidente_por_estado
```

```python
#==========================================
# Pergunta 2
# Meses com maior número de ocorrências de acidentes
#==========================================

gold_acidente_por_mes = (
    df_gold.groupBy("mes_acidente", "nome_mes")
    .agg(count("*").alias("total_acidentes"))
    .orderBy(col("mes_acidente"))
)

print("ACIDENTES POR MÊS")
gold_acidente_por_mes.show()
```

```python
#====================================
# Salvar tabela gold mês
#====================================

gold_acidente_por_mes.write\
    .format("delta")\
    .mode("overwrite")\
    .saveAsTable("projeto_profissional_bigdata.datatran_2018.gold_acidente_por_mes")
```

```python
#================================
# Pergunta 3
# tipos de acidentes frequentes
#================================

gold_tipos_acidentes = (
    df_gold.groupBy("tipo_acidente")
    .agg(count("*").alias("total_acidentes"))
    .orderBy(col("total_acidentes").desc())
)

print("TIPOS DE ACIDENTES MAIS FREQUENTES")
display(gold_tipos_acidentes)
```

```python
gold_tipos_acidentes.write\
    .format("delta")\
    .mode("overwrite")\
    .saveAsTable("projeto_profissional_bigdata.datatran_2018.gold_tipo_acidentes")
```
